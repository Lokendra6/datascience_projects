```python
x=df_new.iloc[:,1:13]
```

```python
y=df_new["AveragePrice"]
```

```python
#Building the model using linearregression and using r2_score for evaluating the model
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_new=scale.fit_transform(x)
```

```python
##We will try to find best random state providing maximum r2_score
```

```python
Maxr2=0
BestRs=0
for i in range(0,200):
    xtrain,xtest,ytrain,ytest=train_test_split(x_new,y,test_size=.25,random_state=i)
    lr=LinearRegression()
    lr.fit(xtrain,ytrain)
    pred=lr.predict(xtest)
    r2=r2_score(pred,ytest)
    if r2>Maxr2:
        Maxr2=r2
        BestRs=i
print("with random state as",BestRs,"max accuracy is",Maxr2)
```

```python
##r2_score is maximum at random state 76 which we will use to build linear model
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_new,y,test_size=.25,random_state=76)
lr.fit(xtrain,ytrain)
lr.score(xtrain,ytrain)
```

```python
pred=lr.predict(xtest)
r2=r2_score(pred,ytest)
print(r2)
```

```python
pred_train=lr.predict(xtrain)
train_accuracy=r2_score(pred_train,ytrain)
print(train_accuracy)
```

```python
test_accuracy=r2_score(pred,ytest)
print(test_accuracy*100)
```

```python
from sklearn.model_selection import cross_val_score
cval=cross_val_score(lr,x_new,y,cv=5)
cv_mean=cval.mean()
print(cv_mean)
```

```python
##Regularization-Using regularization on dataset
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
```

```python
parameter={"alpha":[.0001,.001,.01,.1,1,10],"random_state":list(range(1,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameter)
clf.fit(xtrain,ytrain)
print(clf.best_params_)
print(clf.best_score_)
```

```python
ls=Lasso(alpha=0.0001,random_state=1)
ls.fit(xtrain,ytrain)
ls.score(xtrain,ytrain)
```

```python
predls=ls.predict(xtest)
lss=r2_score(ytest,predls)
lss
```

```python
cval=cross_val_score(ls,x_new,y,cv=5)
cross_score=cval.mean()
print(cross_score)
```

```python
#Using support vector regressor to build model
from sklearn.svm import SVR
```
