```python
y_tmax=df['Next_Tmax']
y_tmin=df["Next_Tmin"]
```

```python
x=df.drop(columns=['Date','Next_Tmax','Next_Tmin'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_new=scale.fit_transform(x)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Identifying the best random state for Next_Tmin
#using the features(x_new) and target variable(y_tmin)
Maxr2_min=0
BestRs_min=0
for i in range(1,200):
    xtrain_tmin,xtest_tmin,ytrain_tmin,ytest_tmin=train_test_split(x_new,y_tmin,test_size=.20,random_state=i)
    lr=LinearRegression()
    lr.fit(xtrain_tmin,ytrain_tmin)
    pred=lr.predict(xtest_tmin)
    accu_min=r2_score(pred,ytest_tmin)
    if accu_min>Maxr2_min:
        Maxr2_min=accu_min
        BestRs_min=i
print("with random state as",BestRs_min,"max accuracy is",Maxr2_min)
```

```python
#Best random state for Next_Tmin is 171
```

```python
#Identifying the best random state for Next_Tmax
#using the features(x_new) and target variable(y_tmax)
Maxr2_max=0
BestRs_max=0
for i in range(1,200):
    xtrain_tmax,xtest_tmax,ytrain_tmax,ytest_tmax=train_test_split(x_new,y_tmax,test_size=.20,random_state=i)
    lr=LinearRegression()
    lr.fit(xtrain_tmax,ytrain_tmax)
    pred=lr.predict(xtest_tmax)
    accu_max=r2_score(pred,ytest_tmax)
    if accu_max>Maxr2_max:
        Maxr2_max=accu_max
        BestRs_max=i
print("with random state as",BestRs_max,"max accuracy is",Maxr2_max)
```

```python
#Best random state for Next_Tmax is 56
```

```python
#Now fitting the model using random state 56 for Next_Tmax
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

xtrain_max,xtest_max,ytrain_max,ytest_max=train_test_split(x_new,y_tmax,test_size=.20,random_state=56)
lr=LinearRegression()
lr.fit(xtrain_max,ytrain_max)
pred=lr.predict(xtest_max)
print("r_score for linear regression",r2_score(pred,ytest_max))
print("mean_squared_error for linear regression",mean_squared_error(pred,ytest_max))
print("mean_absolute_error for linear regression",mean_absolute_error(pred,ytest_max))
print("\n")
print("\n")
rfr=RandomForestRegressor()
rfr.fit(xtrain_max,ytrain_max)
pred=rfr.predict(xtest_max)
print("r_score for RandomForestRegressor",r2_score(pred,ytest_max))
print("mean_squared_error RandomForestRegressor",mean_squared_error(pred,ytest_max))
print("mean_absolute_error for RandomForestRegressor",mean_absolute_error(pred,ytest_max))
print("\n")
print("\n")
svr=SVR()
svr.fit(xtrain_max,ytrain_max)
pred=svr.predict(xtest_max)
print("r_score for Supportvector Regressor",r2_score(pred,ytest_max))
print("mean_squared_error for Supportvector Regressor",mean_squared_error(pred,ytest_max))
print("mean_absolute_error for Supportvector Regressor",mean_absolute_error(pred,ytest_max))
print("\n")
print("\n")
dtr=DecisionTreeRegressor()
dtr.fit(xtrain_max,ytrain_max)
pred=dtr.predict(xtest_max)
print("r_score for DecisionTreeRegressor",r2_score(pred,ytest_max))
print("mean_squared_error for DecisionTreeRegressor",mean_squared_error(pred,ytest_max))
print("mean_absolute_error for DecisionTreeRegressor",mean_absolute_error(pred,ytest_max))

```

```python
#using cross validation to check for over-fitting for target variable(Next_Tmax)
from sklearn.model_selection import cross_val_score
model=[lr,rfr,svr,dtr]
for model in model:
    print(cross_val_score(model,x_new,y_tmax,cv=5).mean())
```

```python
#Now fitting the model using random state 56 for Next_Tmin
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

xtrain_min,xtest_min,ytrain_min,ytest_min=train_test_split(x_new,y_tmin,test_size=.20,random_state=171)
lr=LinearRegression()
lr.fit(xtrain_min,ytrain_min)
pred=lr.predict(xtest_min)
print("r_score for linear regression",r2_score(pred,ytest_min))
print("mean_squared_error for linear regression",mean_squared_error(pred,ytest_min))
print("mean_absolute_error for linear regression",mean_absolute_error(pred,ytest_min))
print("\n")
print("\n")
rfr=RandomForestRegressor()
rfr.fit(xtrain_min,ytrain_min)
pred=rfr.predict(xtest_min)
print("r_score for RandomForestRegressor",r2_score(pred,ytest_min))
print("mean_squared_error RandomForestRegressor",mean_squared_error(pred,ytest_min))
print("mean_absolute_error for RandomForestRegressor",mean_absolute_error(pred,ytest_min))
print("\n")
print("\n")
svr=SVR()
svr.fit(xtrain_min,ytrain_min)
pred=svr.predict(xtest_min)
print("r_score for Supportvector Regressor",r2_score(pred,ytest_min))
print("mean_squared_error for Supportvector Regressor",mean_squared_error(pred,ytest_min))
print("mean_absolute_error for Supportvector Regressor",mean_absolute_error(pred,ytest_min))
print("\n")
print("\n")
dtr=DecisionTreeRegressor()
dtr.fit(xtrain_min,ytrain_min)
pred=dtr.predict(xtest_min)
print("r_score for DecisionTreeRegressor",r2_score(pred,ytest_min))
print("mean_squared_error for DecisionTreeRegressor",mean_squared_error(pred,ytest_min))
print("mean_absolute_error for DecisionTreeRegressor",mean_absolute_error(pred,ytest_min))
```

```python
from sklearn.model_selection import cross_val_score
model=[lr,rfr,svr,dtr]
for model in model:
    print(cross_val_score(model,x_new,y_tmin,cv=5).mean())
```

```python
##for both y_tmax(Next_Tmax) and y_tmin(Next_Tmin),linear regression is the best model to proceed
##from r2_score and cross_val_score it can be deducted that linear regression is the best model to proceed
##As difference between r2_score and cross_val_score is minimum in case of linear regression
```

```python
#Now parameter tuning for y_tmax(Next_Tmax) using GridSearchCV
import numpy as np
from sklearn.model_selection import GridSearchCV
parameter={"fit_intercept":[True,False],"normalize":[True,False],"n_jobs":np.arange(2,10),"positive":[True,False]}
grid=GridSearchCV(estimator=lr,param_grid=parameter,cv=5)
grid.fit(xtrain_max,ytrain_max)
print(grid.best_score_)
print(grid.best_params_)
```

```python
#Now parameter tuning for y_tmin(Next_Tmin) using GridSearchCV
from sklearn.model_selection import GridSearchCV
parameter={"fit_intercept":[True,False],"normalize":[True,False],"n_jobs":np.arange(2,10),"positive":[True,False]}
grid=GridSearchCV(estimator=lr,param_grid=parameter,cv=5)
grid.fit(xtrain_min,ytrain_min)
print(grid.best_score_)
print(grid.best_params_)
```
