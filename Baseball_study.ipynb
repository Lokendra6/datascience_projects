```python
x=df_new.drop("W",axis=1)
y=df_new["W"]
```

```python
df_new.describe()
```

```python
x.shape
```

```python
df_new.columns
```

```python
from  sklearn.preprocessing import power_transform
x_new=power_transform(x)
x_new=pd.DataFrame(x_new,columns=['R', 'AB', 'H', '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'RA', 'ER','ERA', 'CG', 'SHO', 'SV', 'E'])
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scale=scale.fit_transform(x_new)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
lr=LinearRegression()
```

```python
Maxr2=0
BestRs=0
for i in range(0,200):
    xtrain,xtest,ytrain,ytest=train_test_split(x_scale,y,test_size=.25,random_state=i)
    lr=LinearRegression()
    lr.fit(xtrain,ytrain)
    pred=lr.predict(xtest)
    r2=r2_score(pred,ytest)
    if r2>Maxr2:
        Maxr2=r2
        BestRs=i
print("with random state as",BestRs,"max accuracy is",Maxr2)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_scale,y,test_size=.25,random_state=159)
```

```python
lr.fit(xtrain,ytrain)
lr.score(xtrain,ytrain)
```

```python
pred=lr.predict(xtest)
r2=r2_score(pred,ytest)
print(r2)
```

```python
##Cross Validation
pred_train=lr.predict(xtrain)
train_accuracy=r2_score(pred_train,ytrain)
print(train_accuracy)
```

```python
test_accuracy=r2_score(pred,ytest)
print(test_accuracy*100)
```

```python
from sklearn.model_selection import cross_val_score
```

```python

cval=cross_val_score(lr,x_scale,y,cv=5)
cv_mean=cval.mean()
print(cv_mean)
```

```python
##Regularization
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
```

```python
parameter={"alpha":[.0001,.001,.01,.1,1,10],"random_state":list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameter)
clf.fit(xtrain,ytrain)
```

```python
print(clf.best_params_)
```

```python
print(clf.best_score_)
```

```python
ls=Lasso(alpha=1,random_state=0)
ls.fit(xtrain,ytrain)
ls.score(xtrain,ytrain)
```

```python
predls=ls.predict(xtest)
lss=r2_score(ytest,predls)
lss
```

```python
cval=cross_val_score(ls,x_scale,y,cv=5)
cross_score=cval.mean()
print(cross_score)
```

```python
##emsemble technique
from sklearn.ensemble import RandomForestRegressor
parameters={"criterion":["mse","mae"],"max_features":["auto","sqrt","log2"]}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(xtrain,ytrain)
```

```python
print(clf.best_params_)
```

```python
print(clf.best_score_)
```

```python
rf=RandomForestRegressor(criterion="mse",max_features="auto" )
rf.fit(xtrain,ytrain)
rf.score(xtrain,ytrain)
```

```python
predrf=rf.predict(xtest)
rfs=r2_score(predrf,ytest)
print(rfs)
```

```python
cvalrf=cross_val_score(rf,x_scale,y,cv=5)
cross_score=cvalrf.mean()
print(cross_score)
```
